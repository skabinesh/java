### 6.Write a Python program using Pandas to calculate various probabilities from the sales dataset. Compute the marginal probability of a customer purchasing a Laptop, the marginal probability of a customer being from the North region, and the joint probability of both events occurring together. Then, calculate the conditional probability of a customer being from the North region given they purchased a Laptop, and finally compute the union probability of either event occurring. Display all results rounded to two decimal places.

```python
# Marginal probability: P(Product = Laptop)
prob_laptop = (df['Product'] == 'Laptop').mean()

# Joint probability: P(Product = Laptop and Region = North)
prob_joint = ((df['Product'] == 'Laptop') & (df['Region'] == 'North')).mean()

# Marginal for Region = North
prob_north = (df['Region'] == 'North').mean()

# Conditional probability: P(Region = North | Product = Laptop)
prob_cond = prob_joint / prob_laptop

# Union probability: P(Laptop or North) = P(A) + P(B) - P(A and B)
prob_union = prob_laptop + prob_north - prob_joint

print("P(Laptop):", round(prob_laptop, 2))
print("P(North):", round(prob_north, 2))
print("P(Laptop ∩ North):", round(prob_joint, 2))
print("P(North | Laptop):", round(prob_cond, 2))
print("P(Laptop ∪ North):", round(prob_union, 2))
```

### 7. Write a Python program using Pandas and Matplotlib to perform a time-series analysis of monthly sales data. Resample the dataset to calculate total monthly revenue based on the ‘Date’ column, and plot the resulting trend using a line chart with markers. Add appropriate axis labels, a title, and grid lines to enhance readability.

### 8.Write a Python program using Pandas and SciPy to perform a Chi-Squared test to examine the relationship between customer demographics and product preferences. Create a contingency table between 'Sales_Rep' and 'Product' columns to simulate demographics and preferences, then perform the Chi-Squared test. Display the Chi-squared statistic and p-value rounded appropriately, and interpret whether the relationship is statistically significant.

```python
from scipy.stats import chi2_contingency

# Cross-tab between Sales Rep and Product (simulated demographic vs preference)
contingency_table = pd.crosstab(df['Sales_Rep'], df['Product'])
contingency_table
# Chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)

print("Chi-squared value:", round(chi2, 2))
print("P-value:", round(p, 4))
if p < 0.05:
    print("There is a significant relationship between Sales Rep and Product preference.")
else:
    print("No significant relationship found.")
```

### 9. Write a Python program using NumPy, SciPy, and Matplotlib to simulate and visualize a Poisson distribution representing customer arrivals. Assume the average number of arrivals (λ) is 4 per hour. Compute the probability mass function for 0 to 10 customers arriving in an hour and plot a bar graph showing the probabilities with appropriate labels, title, and grid lines.


```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Simulate Poisson distribution
x = np.arange(0, 11)  # possible number of arrivals
lambda_val = 4
probs = poisson.pmf(x, mu=lambda_val)

# Plot the distribution
plt.figure(figsize=(6, 4))
plt.bar(x, probs, color='lightgreen')
plt.title('Poisson Distribution (Customer Arrivals per Hour)')
plt.xlabel('Number of Customers')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

### 10. Write a Python program using SciPy to calculate the probability of selecting exactly 2 defective products from a batch of 20, where 5 products are known to be defective. Assume a random sample of 6 products is selected without replacement. Use the hypergeometric distribution to compute the probability and display the result rounded to four decimal places.

```python
from scipy.stats import hypergeom

# Parameters: M = population size, n = total defectives, N = sample size
M = 20  # total products
n = 5   # defective products
N = 6   # sample taken

# Probability of getting exactly 2 defective items
prob_2_defectives = hypergeom.pmf(2, M, n, N)

print("P(Exactly 2 defectives in sample):", round(prob_2_defectives, 4))
```

### 12. Write a Python program using Pandas to create a DataFrame of students with their scores in Math, Science, and English. Add a new column for average marks. Filter the students who scored more than 75 in all subjects, sort the DataFrame by average in descending order, and save the result to a CSV file.

```python
import pandas as pd

data = {
    'Name': ['Amit', 'Sana', 'Raj', 'Pooja', 'Karan'],
    'Math': [85, 72, 78, 88, 95],
    'Science': [91, 79, 69, 92, 89],
    'English': [76, 81, 74, 90, 86]
}

df = pd.DataFrame(data)
df['Average'] = df[['Math', 'Science', 'English']].mean(axis=1)
filtered_df = df[(df['Math'] > 75) & (df['Science'] > 75) & (df['English'] > 75)]
sorted_df = filtered_df.sort_values(by='Average', ascending=False)
sorted_df.to_csv("top_students.csv", index=False)
print(sorted_df)
```

### 13.Write a Python program using NumPy to create a 1D array of 24 random integers between 10 and 100. Reshape it to a 4x6 2D array, find the sum row-wise and column-wise, extract even numbers, and flatten the array back to 1D.

```python
import numpy as np

arr = np.random.randint(10, 100, 24)
arr_2d = arr.reshape(4, 6)
print("2D Array:\n", arr_2d)

row_sum = arr_2d.sum(axis=1)
col_sum = arr_2d.sum(axis=0)
print("Row-wise Sum:", row_sum)
print("Column-wise Sum:", col_sum)

even_numbers = arr_2d[arr_2d % 2 == 0]
print("Even Numbers:", even_numbers)

flattened = arr_2d.flatten()
print("Flattened Array:", flattened)
```

### 15.Write a Python program using Pandas to create a DataFrame with Product, Price, and Quantity. Add a new column called Total (Price * Quantity). Filter products with Total greater than 500. Sort the DataFrame by Total and save the result to an Excel file.

```python
import pandas as pd

data = {
    'Product': ['Pen', 'Notebook', 'Bag', 'Bottle', 'Pencil'],
    'Price': [20, 50, 400, 120, 10],
    'Quantity': [10, 5, 2, 3, 25]
}

df = pd.DataFrame(data)
df['Total'] = df['Price'] * df['Quantity']
filtered = df[df['Total'] > 500]
sorted_df = filtered.sort_values(by='Total', ascending=False)
sorted_df.to_excel("high_value_products.xlsx", index=False)
print(sorted_df)
```
